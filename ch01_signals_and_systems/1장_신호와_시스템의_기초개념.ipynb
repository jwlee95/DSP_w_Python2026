## 1.1 신호의 정의와 분류 (Definition and Classification of Signals)

### 1.1.1 신호란 무엇인가 (What Is a Signal)

신호(signal)는 시간이 지나면서 변화하는 어떤 값을 함수의 형태로 표현한 것이다. 하루 동안 한 시간마다 측정한 체온 기록이나, 마이크가 매 순간 감지하는 공기압 변화 모두 신호의 한 예이다.

신호는 독립변수인 시간이 어떻게 흐르는가에 따라 두 가지로 구분된다.

- **연속시간 신호(continuous-time signal)** $x(t)$: 시간 $t$가 끊김 없이 흐르는 동안 정의되는 신호이다. 수은 온도계가 보여주는 온도 변화가 그 예이다.
- **이산시간 신호(discrete-time signal)** $x[n]$: 시간을 일정한 간격으로 나누어, 그 순간의 값만을 기록한 신호이다. 한 시간마다 한 번씩 측정한 체온 기록표가 그 예이며, 여기서 $n$은 몇 번째 측정인지를 나타내는 정수(0, 1, 2, …)일 뿐이다.

이 두 가지 신호의 구분은 신호처리 분야에서 가장 기본이 되는 표준적인 분류 방식이다.

> $x(t)$와 $x[n]$이 정확히 어떤 수식적 관계로 연결되는지(즉 $t$가 $n$으로 어떻게 바뀌는지)는 1.3.3절에서 자세히 다룬다.

### 1.1.2 신호의 분류 (Classification of Signals)

신호를 분류하는 목적은, 신호의 성질에 따라 적용되는 수학적 도구가 달라지기 때문이다. 학부 단계에서는 다음 세 가지 분류 기준을 직관적으로 이해하는 것으로 충분하다.

| 분류 기준 | 직관적 의미 | 수식적 정의 |
|---|---|---|
| 주기/비주기 | 동일한 패턴이 일정한 시간마다 반복되는가 (예: 시계 초침 소리는 주기적, 사람의 말소리는 비주기적이다) | $x(t) = x(t+T_0)$ |
| 에너지/전력 신호 | 신호의 총량이 유한한가(에너지 신호, 예: 한 번 친 종소리), 무한히 지속되는가(전력 신호, 예: 라디오 방송) | $E_x = \displaystyle\int_{-\infty}^{\infty} \lvert x(t)\rvert^2\,dt$ |
| 짝/홀신호 | 그래프를 $y$축 기준으로 좌우 대칭으로 접었을 때 겹치는가(짝신호), 위아래로 뒤집힌 모양으로 겹치는가(홀신호) | $x(-t)=x(t)$ 또는 $x(-t)=-x(t)$ |

### 1.1.3 대표 예제 신호: 사인파 (Representative Example Signal: Sinusoid)

본 교재의 본문 예제로 가장 빈번하게 사용되는 신호는 사인파(정현파)이다.

$$
x(t) = A\cos(\omega_0 t + \phi) \tag{1.1}
$$

이 수식을 구성하는 세 파라미터의 역할은 다음과 같이 이해할 수 있다.

- $A$ (진폭, amplitude): 신호가 위아래로 흔들리는 크기, 즉 그래프의 높이를 결정한다.
- $\omega_0 = 2\pi f_0$ (각주파수, angular frequency): 1초 동안 몇 번 흔들리는가를 나타내며, $f_0$는 통상적으로 사용하는 주파수(단위: Hz)이다. 이 값이 클수록 신호가 빠르게 진동한다.
- $\phi$ (위상, phase): 신호가 시간축 상에서 좌우로 얼마나 이동되어 있는지를 나타낸다.

## 1.2 시스템의 정의 (Definition of a System)

시스템(system)은 입력 신호를 받아 출력 신호로 변환하는 연산자(operator)로 정의된다.

$$
y(t) = T[x(t)] \tag{1.2}
$$

오디오 이퀄라이저(equalizer)는 시스템의 한 예이다. 마이크에서 들어온 소리(입력 $x(t)$)를 받아 저음을 강조하고 고음을 줄이는 처리를 거친 뒤, 스피커로 내보낼 소리(출력 $y(t)$)를 만들어낸다. 여기서 $T[\cdot]$는 그 처리 과정 전체를 하나의 기호로 표현한 것이다.

본 장에서는 "시스템은 입력과 출력을 연결하는 규칙"이라는 수준까지의 이해를 목표로 한다. 그 규칙이 갖추어야 할 구체적인 성질(선형성(linearity), 시불변성(time-invariance), 인과성(causality), 안정성(stability))은 2장과 3장에서 본격적으로 다룬다.

## 1.3 디지털 신호처리(DSP)의 개념과 장단점 (Concept and Trade-offs of Digital Signal Processing)

### 1.3.1 아날로그-디지털 변환의 구조 (Structure of Analog-to-Digital Conversion)

자연계의 신호(소리, 온도, 진동 등)는 본질적으로 연속시간 신호, 즉 아날로그 신호이다. 그러나 컴퓨터는 숫자만을 처리할 수 있으므로, 아날로그 신호를 숫자들의 나열인 디지털 신호로 변환하는 장치가 필요하다. 이 역할을 수행하는 것이 A/D 변환기(ADC)이며, 처리가 끝난 디지털 결과를 다시 아날로그 신호로 되돌리는 장치가 D/A 변환기(DAC)이다.

$$
x(t) \;\xrightarrow{\text{A/D 변환}}\; x[n] \;\xrightarrow{\text{디지털 처리}}\; y[n] \;\xrightarrow{\text{D/A 변환}}\; y(t) \tag{1.3}
$$

### 1.3.2 디지털 신호처리의 장단점 (Advantages and Disadvantages of DSP)

| 장점 | 단점 |
|---|---|
| 재현성이 높다 (복제해도 품질이 유지된다) | 표본화(sampling) 과정에서 정보가 손실될 수 있다 |
| 잡음에 대한 강인성이 비교적 높다 | 양자화(quantization) 과정에서 오차가 발생한다 |
| 소프트웨어 변경만으로 처리 방식을 유연하게 바꿀 수 있다 | 변환 및 연산 과정에서 지연이 발생한다 |
| 저장 및 전송이 용이하다 | 하드웨어(ADC, DAC, 프로세서) 비용이 수반된다 |

표본화 과정에서 발생하는 정보 손실(앨리어싱, aliasing)과 나이퀴스트 조건(Nyquist condition)에 대한 엄밀한 논의는 4장(샘플링 이론)에서 다루며, 본 장에서는 개념적 예고 수준으로 제한한다.

### 1.3.3 표기법의 변환: $x(t) \to x[n]$, $\omega \to \Omega$

**(1) 괄호 표기의 관행**

연속시간 신호는 소괄호 $x(t)$, 이산시간 신호는 대괄호 $x[n]$으로 표기하는 것이 표준적인 관행(convention)이다.

- $t$: 실수 전체에서 끊김 없이 변하는 연속적인 독립변수 (단위: 초)
- $n$: 0, 1, 2, …와 같이 정수만을 취하는 이산적인 독립변수 (단위 없음, 단지 "몇 번째"를 나타내는 인덱스)

**(2) 시간축의 변환: $t \to n$**

연속신호 $x(t)$를 표본화 주파수 $f_s$로 표본화하면, 표본화 간격 $T_s = 1/f_s$마다 한 번씩 측정한 값들이 이산신호가 된다.

$$
t = nT_s = \frac{n}{f_s} \tag{1.4}
$$

$$
x[n] \;\triangleq\; x(t)\Big|_{t = n/f_s} \;=\; x(nT_s) \tag{1.5}
$$

**(3) 각주파수의 변환: $\omega \to \Omega$**

$$
x[n] = x(t)\Big|_{t=nT_s} = A\cos(\omega_0\, n T_s + \phi) = A\cos\!\big(\underbrace{\omega_0 T_s}_{\displaystyle \Omega_0}\, n + \phi\big) = A\cos(\Omega_0 n + \phi) \tag{1.6}
$$

$$
\Omega_0 \;\triangleq\; \omega_0 T_s \;=\; \frac{\omega_0}{f_s} \;=\; \frac{2\pi f_0}{f_s} \tag{1.7}
$$

이를 **이산시간 각주파수(discrete-time angular frequency)**라 부르며, 통상적으로 대문자 $\Omega$로 표기하여 연속시간 각주파수 $\omega$와 구분한다.

| | $\omega_0$ (연속시간 각주파수) | $\Omega_0$ (이산시간 각주파수) |
|---|---|---|
| 단위 | rad/s (초당 라디안) | rad/sample (표본당 라디안) |
| 의미 | 실제 물리적 시간에 대한 변화율 | 표본화 주파수로 정규화(normalize)된, 표본 인덱스에 대한 변화율 |
| 값의 결정 요인 | 신호 자체의 물리적 주파수 $f_0$ 만으로 결정 | $f_0$ 뿐만 아니라 표본화 주파수 $f_s$에도 의존 |

**수치 예시**: $f_0 = 5$ Hz인 신호를 $f_s = 50$ Hz로 표본화하면 $\Omega_0 = 2\pi(5)/50 = \pi/5 \approx 0.628$ rad/sample이다. 같은 신호를 $f_s = 20$ Hz로 표본화한다면 $\Omega_0 = 2\pi(5)/20 = \pi/2 \approx 1.571$ rad/sample로, 표본화 주파수에 따라 값이 달라짐을 확인할 수 있다.

$\cos(\cdot)$ 함수는 $2\pi$를 주기로 반복되므로 $\Omega_0$는 본질적으로 $2\pi$를 모듈로(modulo)로 하는 양이며, 통상 $[-\pi, \pi)$ 또는 $[0, 2\pi)$ 구간에서만 고유한 의미를 가진다.

> **구간 표기 $[-\pi, \pi)$에 대하여**: 대괄호 `[`와 소괄호 `)`가 함께 쓰인 것은, "한쪽 끝점은 포함하고 다른 쪽 끝점은 포함하지 않는 구간(반개구간, half-open interval)"을 나타내는 표준적인 수학 표기이다. $\Omega_0 = -\pi$와 $\Omega_0 = \pi$가 $\cos(\cdot)$ 함수에서 정확히 같은 값을 만들어내므로($\cos(-\pi n) = \cos(\pi n)$), 중복을 피하기 위해 구간의 한쪽 끝만을 대표값으로 포함시킨다.

이는 연속신호에는 없는, 이산신호만의 독특한 성질이며 그 구체적인 함의는 6장(DTFT)에서 본격적으로 다룬다.

## 1.4 기초 신호 (Elementary Signals)

이산신호처리에서 반복적으로 사용되는 기본 신호들을 정의한다. 복잡한 신호도 결국 이 기본 신호들의 조합으로 표현될 수 있다는 점이 핵심이다.

- **단위 임펄스(unit impulse)** $\delta[n]$:
$$
\delta[n] = \begin{cases}1, & n=0\\0, & n\neq0\end{cases} \tag{1.8}
$$
- **단위 계단(unit step)** $u[n]$:
$$
u[n] = \begin{cases}1, & n\geq0\\0, & n<0\end{cases} \tag{1.9}
$$
- **지수신호(exponential signal)**: $x[n] = a^n$
- **정현파(sinusoid)**: $x[n] = A\cos(\Omega_0 n + \phi)$

이 기본 신호들은 이후 3장(컨볼루션)과 8장(z변환)을 전개하는 출발점이 된다.

## 1.5 핵심 알고리즘: 연속신호의 샘플링 시뮬레이션

$$
\begin{array}{ll}
\textbf{Algorithm 1.1} & \text{연속신호의 샘플링 시뮬레이션 (Sampling simulation of a continuous-time signal)} \\[6pt]
\textbf{Require:} & f_0 \text{ (신호 주파수)},\ f_s \text{ (표본화 주파수)},\ T_{\text{dur}} \text{ (측정 시간)} \\
\textbf{Ensure:} & (\mathbf{t}_c, \mathbf{x}_c) \text{ (연속신호 근사)},\ (\mathbf{t}_s, \mathbf{x}_s) \text{ (표본화된 이산신호)} \\[6pt]
1: & \mathbf{t}_c \leftarrow \text{구간 } [0, T_{\text{dur}}] \text{ 을 매우 촘촘하게 등분한 시간 벡터} \\
2: & \textbf{for } t \in \mathbf{t}_c \textbf{ do} \\
3: & \quad x_c(t) \leftarrow A\cos(2\pi f_0 t + \phi) \\
4: & \textbf{end for} \\
5: & N \leftarrow \lfloor f_s \cdot T_{\text{dur}} \rfloor \\
6: & \textbf{for } n = 0 \textbf{ to } N \textbf{ do} \\
7: & \quad t_s[n] \leftarrow n / f_s \\
8: & \quad x_s[n] \leftarrow A\cos(2\pi f_0\, t_s[n] + \phi) \\
9: & \textbf{end for} \\
10: & \textbf{return } (\mathbf{t}_c, \mathbf{x}_c, \mathbf{t}_s, \mathbf{x}_s)
\end{array}
$$

- **입력(Require)**: 신호 주파수 $f_0$, 표본화 주파수 $f_s$, 측정 시간 $T_{\text{dur}}$를 입력으로 받는다.
- **1행**: 시간축 $\mathbf{t}_c$를 구간 $[0, T_{\text{dur}}]$에서 매우 잘게 나눈 벡터로 생성한다.
- **2~4행**: $\mathbf{t}_c$의 각 시간 $t$에 사인파 수식을 대입하여 연속신호 근사 곡선 $x_c(t)$를 만든다.
- **5행**: 표본의 총 개수 $N$을 표본화 주파수와 측정 시간의 곱으로 계산한다. $\lfloor \cdot \rfloor$는 바닥함수(floor function)로서, 파이썬에서는 `int(...)` 또는 `//` 연산자가 그 역할을 수행한다.
- **6~9행**: $n=0$부터 $N$까지의 각 정수 인덱스에 대해 표본 시각 $t_s[n]=n/f_s$를 계산하고, 동일한 사인파 수식으로 이산신호 $x_s[n]$을 계산한다.
- **출력(Return)**: 근사 곡선 $(\mathbf{t}_c, \mathbf{x}_c)$와 표본점 $(\mathbf{t}_s, \mathbf{x}_s)$을 함께 반환한다.

## 1.6 파이썬 구현 (Python Implementation)

아래 코드는 세 개의 셀로 나누어 구성하였다. 첫 번째 셀은 신호 생성 함수, 두 번째 셀은 시각화 함수, 세 번째 셀은 실행부이다. **본 노트북은 결과를 미리 포함하지 않으므로, 아래 셀들을 순서대로 직접 실행하여 결과를 확인하기 바란다.**


```python
"""
1장 예제: 연속신호 근사 곡선과 표본화된 이산신호를 함께 생성한다.
필요 라이브러리: numpy, matplotlib
"""

import numpy as np
import matplotlib.pyplot as plt


def generate_continuous_and_sampled_sine(A: float, f0: float, phi: float,
                                          fs: float, duration: float,
                                          n_cont: int = 2000):
    """
    사인파 x(t) = A*cos(2*pi*f0*t + phi)에 대하여,
    연속신호 근사 곡선과 fs[Hz]로 표본화한 이산신호를 동시에 생성한다.

    Parameters
    ----------
    A : 진폭(amplitude)
    f0 : 신호 주파수(Hz)
    phi : 위상(rad)
    fs : 표본화 주파수(sampling frequency, Hz)
    duration : 측정 시간(s)
    n_cont : 연속신호 근사를 위한 점의 개수 (기본값 2000)

    Returns
    -------
    t_cont, x_cont : 연속신호 근사를 위한 시간 벡터와 신호 값
    t_samp, x_samp : 표본화된 이산신호의 시간 벡터와 신호 값
    """
    t_cont = np.linspace(0, duration, n_cont)
    x_cont = A * np.cos(2 * np.pi * f0 * t_cont + phi)

    n = np.arange(0, int(fs * duration) + 1)
    t_samp = n / fs
    x_samp = A * np.cos(2 * np.pi * f0 * t_samp + phi)

    return t_cont, x_cont, t_samp, x_samp
```

```python
def plot_continuous_vs_sampled(t_cont, x_cont, t_samp, x_samp, f0, fs):
    """연속신호 근사 곡선과 표본점을 흰색 배경의 그래프 위에 함께 그린다."""
    fig, ax = plt.subplots(figsize=(8, 4), facecolor='white')
    ax.set_facecolor('white')

    ax.plot(t_cont, x_cont, color='black', linewidth=1.5,
            label='x(t), continuous')
    ax.stem(t_samp, x_samp, linefmt='b-', markerfmt='bo', basefmt=' ',
            label='x[n], sampled')

    ax.set_xlabel('Time t (s)')
    ax.set_ylabel('Amplitude x(t)')
    ax.set_title(f'Sine wave (f0 = {f0} Hz) sampled at fs = {fs} Hz')
    ax.legend(loc='upper right')
    ax.grid(True, linewidth=0.5, alpha=0.5)

    fig.tight_layout()
    plt.show()
```

```python
# 실행부: 파라미터를 지정하고 함수를 호출한다.
A, f0, phi = 1.0, 5.0, 0.0      # 진폭, 주파수(Hz), 위상(rad)
fs, duration = 50.0, 1.0        # 표본화 주파수(Hz), 측정 시간(s)

t_cont, x_cont, t_samp, x_samp = generate_continuous_and_sampled_sine(
    A, f0, phi, fs, duration
)

plot_continuous_vs_sampled(t_cont, x_cont, t_samp, x_samp, f0, fs)
```

### 코드 해설

1. **모듈 docstring**: 이 스크립트가 무엇을 하는지, 실행을 위해 어떤 라이브러리가 필요한지를 가장 먼저 명시한다.
2. **`import numpy as np`**: 배열(array) 형태의 수치 연산을 빠르게 수행하는 라이브러리이며, 본 교재 전반에서 가장 기본이 되는 도구이다.
3. **`import matplotlib.pyplot as plt`**: 그래프 작성을 위한 라이브러리이다.
4. **`generate_continuous_and_sampled_sine(...)`**: 신호 생성 부분을 별도의 함수로 분리하였다.
   - `t_cont = np.linspace(0, duration, n_cont)`: 0초부터 `duration`초까지를 `n_cont`개의 점으로 균일하게 나눈 배열을 생성하여, 연속신호에 근접한 시간축을 구성한다.
   - `x_cont = A * np.cos(2 * np.pi * f0 * t_cont + phi)`: `t_cont`가 배열임에도 이 한 줄로 모든 원소에 대한 결과를 동시에 계산한다(벡터화, vectorization).
   - `n = np.arange(0, int(fs * duration) + 1)`: 표본의 인덱스 $n=0,1,2,\dots,N$에 해당하는 정수 배열을 생성한다.
   - `t_samp = n / fs`: 각 인덱스 $n$을 실제 측정 시각(초 단위)으로 환산한다.
   - `x_samp = A * np.cos(2 * np.pi * f0 * t_samp + phi)`: `x_cont`와 동일한 수식이지만, 표본 시각(`t_samp`)에 대해서만 값을 계산한다. 이 결과가 이산신호 $x[n]$에 해당한다.
5. **`plot_continuous_vs_sampled(...)`**: 시각화 부분을 별도의 함수로 분리하였다.
   - `facecolor='white'` 및 `ax.set_facecolor('white')`: 그래프의 배경을 흰색으로 명시적으로 지정한다.
   - `ax.plot(...)`: 연속신호 근사 곡선을 검은색 실선으로 그린다.
   - `ax.stem(...)`: 표본화된 이산신호를 줄기-마커(stem) 형태로 그리며, 이산신호를 표현하는 표준적인 방법이다.
   - `ax.set_xlabel(...)`, `ax.set_ylabel(...)`: 축 라벨을 영어로 지정한다.
6. **실행부**: 파라미터를 지정한 뒤 두 함수를 순서대로 호출한다. 스크립트 형태와 달리 노트북에서는 `if __name__ == "__main__":` 블록 없이 실행부를 별도의 셀로 분리하는 것이 관례이다.

여기서 강조해야 할 점은, `x_cont`와 `x_samp`를 계산하는 수식이 완전히 동일하다는 것이다. 즉 연속신호와 이산신호는 본질적으로 다른 신호가 아니라, 동일한 수식을 얼마나 촘촘하게(또는 듬성듬성하게) 평가하였는가의 차이일 뿐이다. 이것이 1장 전체에서 전달하고자 하는 핵심 직관이다.

## 1.7 시각화와 해석 (Visualization and Interpretation)

연속신호 $x(t)$와 표본화된 이산신호 $x[n]$을 동일한 좌표축 위에 겹쳐 그리면, 검은 실선의 연속신호 위에 일정 간격으로 표본점이 찍힌 형태를 확인할 수 있다. 이 그림을 통해 다음 세 가지를 전달할 수 있다.

1. 이산신호는 연속신호 위의 등간격 표본점이다.
2. 표본화 간격 $T_s = 1/f_s$가 표본점 사이의 시간축 거리를 결정한다.
3. 표본화 주파수가 충분히 크지 않으면(나이퀴스트 조건 미충족) 표본점들로부터 원신호를 복원할 수 없으며, 이는 4장에서 다룰 앨리어싱 현상과 연결된다.

본 교재의 모든 그림은 다음 원칙을 일관되게 따른다.

- 그래프의 배경은 흰색으로 통일한다.
- 축 라벨(`Time t (s)`, `Amplitude x(t)` 등)은 영어로 표기한다.
- 그래프 위의 부가 설명(주석)은 한글로 병기한다.
- 색상은 최소한으로 사용하며, 연속신호는 흑색 실선, 이산신호는 단일 색의 점으로 구분한다.
