# 쌍곡선 함수와 우기 분해 (Even-Odd Decomposition)

## 1. 우-기 분해 (Even-Odd Decomposition)의 기본 개념

세상의 모든 임의의 신호나 함수 $f(x)$는 우함수(짝수 성분)와 기함수(홀수 성분)의 합으로 완벽하게 쪼갤 수 있으며, 이를 `우-기 분해`라고 합니다.

- **우함수 성분 (Even Part):** $y$축을 기준으로 좌우가 대칭인 성분   
    $$
    f_e(x) = \frac{f(x) + f(-x)}{2}
    $$

- **기함수 성분 (Odd Part):** 원점 $(0, 0)$을 기준으로 점대칭인 성분   
    $$
    f_o(x) = \frac{f(x) - f(-x)}{2}
    $$

- **우기 분해의 완성:**   
    $$
    f(x) = f_e(x) + f_o(x)
    $$


## 2. 쌍곡선 함수 (Hyperbolic Functions)의 대칭성

쌍곡선 함수는 삼각함수와 유사한 대칭적 특성을 지니고 있습니다.

- **$\cosh(x)$ (하이퍼볼릭 코사인):** **우함수 (Even)**
    - 특징: $y$축 대칭
    - 수학적 표현:
    $$
    \cosh(-x) = \cosh(x)
    $$
- **$\sinh(x)$ (하이퍼볼릭 사인):** **기함수 (Odd)**
    - 특징: 원점 대칭
    - 수학적 표현:
    $$\sinh(-x) = -\sinh(x)
    $$
- **$\tanh(x)$ (하이퍼볼릭 탄젠트):** **기함수 (Odd)**
    - 특징: $\frac{\sinh(x)}{\cosh(x)}$ 이므로 원점 대칭
    - 수학적 표현:
    $$
    \tanh(-x) = -\tanh(x)
    $$

## 3. 우-기 분해 (Even-Odd Decomposition) 예제 시각화

```python
import numpy as np
import matplotlib.pyplot as plt

# 우-기 분해 함수 정의
def even_odd_decomposition(f, x):
    f_x = f(x)
    f_neg_x = f(-x)
    even_part = (f_x + f_neg_x) / 2
    odd_part = (f_x - f_neg_x) / 2
    return even_part, odd_part

# 예제 함수: f(x) = e^x
def example_function(x):
    return np.exp(x)

# x 값 범위 설정
x_values = np.linspace(-3, 3, 400)

# 원래 함수 계산
y_original = example_function(x_values)

# 우-기 분해 계산
y_even, y_odd = even_odd_decomposition(example_function, x_values)

# 플롯 그리기
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x_values, y_original, label='$f(x) = e^x$', color='blue')
plt.title('Original Function $f(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(x_values, y_even, label='$f_e(x)$ (Even Part)', color='green')
plt.title('Even Part $f_e(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(x_values, y_odd, label='$f_o(x)$ (Odd Part)', color='red')
plt.title('Odd Part $f_o(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

이 그래프는 $f(x) = e^x$ 함수를 우함수 성분 $f_e(x)$와 기함수 성분 $f_o(x)$로 분해한 것을 보여줍니다.

- **Original Function $f(x)$:** 원래의 $e^x$ 함수입니다.
- **Even Part $f_e(x)$:** $y$축 대칭인 우함수 성분입니다. ($e^x$의 우함수 성분은 $\cosh(x)$와 같습니다.)
- **Odd Part $f_o(x)$:** 원점 대칭인 기함수 성분입니다. ($e^x$의 기함수 성분은 $\sinh(x)$와 같습니다.)

확인할 수 있듯이, $f_e(x) + f_o(x)$를 더하면 원래 함수 $f(x)$가 됩니다.

## 4. 우-기 분해 결과 확인: $f_e(x) + f_o(x) = f(x)$

```python
import matplotlib.pyplot as plt

# 우함수 성분과 기함수 성분을 합산
y_sum_even_odd = y_even + y_odd

# 원래 함수와 합산 결과를 플롯
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_original, label='Original Function $f(x)$', color='blue', linestyle='-')
plt.plot(x_values, y_sum_even_odd, label='$f_e(x) + f_o(x)$', color='red', linestyle='--')
plt.title('Verification: $f_e(x) + f_o(x)$ vs $f(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid(True)
plt.show()

# 수치적으로도 일치하는지 확인 (선택 사항)
# print("Maximum absolute difference between f(x) and (fe(x) + fo(x)):", np.max(np.abs(y_original - y_sum_even_odd)))
```

## 5. 핵심: 지수함수 $e^x$의 우기 분해

가장 중요한 포인트는 **지수함수와 쌍곡선 함수의 관계**입니다.

비대칭 함수인 자연상수 지수함수 $f(x) = e^x$를 우기 분해 공식에 대입하면 놀라운 결과가 나타납니다.

### ① $e^x$의 우함수 성분 추출

$$
f_e(x) = \frac{e^x + e^{-x}}{2}
$$

> **💡 이것은 정확히 $\cosh(x)$의 정의와 같습니다.**

### ② $e^x$의 기함수 성분 추출

$$
f_o(x) = \frac{e^x - e^{-x}}{2}
$$

> **💡 이것은 정확히 $\sinh(x)$의 정의와 같습니다.**

### ③ 최종 결론

지수함수는 자신의 우함수 성분과 기함수 성분의 합이므로, 다음과 같은 우아한 수식이 성립합니다.
$$
e^x = \cosh(x) + \sinh(x)
$$
일반 삼각함수의 가장 유명한 공식인 $\cos^2 x + \sin^2 x = 1$ 은 원의 방정식에서 왔듯이, 쌍곡선함수는 쌍곡선의 방정식($x^2 - y^2 = 1$)에서 왔기 때문에 가운데 부호가 마이너스($-$)가 됩니다.
$$
\cosh^2(x) - \sinh^2(x) = 1
$$
*(참고: 이 항등식 양변을 $\cosh^2(x)$로 나누면 $1 - \tanh^2(x) = \text{sech}^2(x)$ 라는 공식도 유도됩니다.)*

또한, 일반 삼각함수를 미분할 때는 $\cos$을 미분하면 $-\sin$이 되는 등 부호가 바뀌어 헷갈리는 경우가 많은데, 쌍곡선함수는 지수함수 $e^x$의 조합으로 이루어져 있어 **미분할 때 부호가 거의 바뀌지 않는다는 아주 편리한 특성**이 있습니다.

- $(\sinh x)' = \cosh x$
- $(\cosh x)' = \sinh x$  **(주의: 마이너스가 붙지 않음!)**
- $(\tanh x)' = \text{sech}^2 x$

## 6. 쌍곡선 함수 ($\cosh(x)$, $\sinh(x)$, $\tanh(x)$) 그래프 시각화

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 범위 설정 (이전 셀에서 정의된 x_values 재사용 가능)
# x_values = np.linspace(-3, 3, 400)

# 쌍곡선 함수 계산
y_cosh = np.cosh(x_values)
y_sinh = np.sinh(x_values)
y_tanh = np.tanh(x_values)

# 플롯 그리기
plt.figure(figsize=(18, 5))

# cosh(x) 플롯
plt.subplot(1, 3, 1)
plt.plot(x_values, y_cosh, label=r'$\cosh(x)$', color='purple')
plt.title(r'Hyperbolic Cosine ($\cosh(x)$)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid(True)

# sinh(x) 플롯
plt.subplot(1, 3, 2)
plt.plot(x_values, y_sinh, label=r'$\sinh(x)$', color='orange')
plt.title(r'Hyperbolic Sine ($\sinh(x)$)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid(True)

# tanh(x) 플롯
plt.subplot(1, 3, 3)
plt.plot(x_values, y_tanh, label=r'$\tanh(x)$', color='brown')
plt.title(r'Hyperbolic Tangent ($\tanh(x)$)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
